```python
#import necessary libraries for data handling and visualisation as matplot ans sns
import pandas as pd
```

```python
#creat data set manually using a dictionary
data = {
    'city':['algiers' , 'oran' , 'setif' , 'algiers' , 'oran' , 'setif'] ,
    'consumption_m3':[12000,15000,11000,13000,14000,10000],
    'breakdowns' : [20,15,18,22,14,17]
}
```

```python
#convert the dictionary into a data frame for analysis
df = pd.DataFrame(data)
```

```python
#preview the dataset
print(df)
```

```python
#averages
print("average consumption :" , df['consumption_m3'].mean())
print("average breakdowns :" , df ['breakdowns'].mean())
```

```python
#correlation linéaire etre les deux variables
print("correlation matrix :")
print(df[['consumption_m3','breakdowns']].corr())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#create a scatter plot to visualise the relashionship
plt.figure(figsize=(8,6)) #the size of the plot
#plot the data points with seaborn and color by city hue
sns.scatterplot(data=df,x='consumption_m3',y='breakdowns',hue = 'city')
#add cities and lables
plt.title("water consumption vs. breakdowns by city")
plt.xlabel("consumption(m³)")
plt.ylabel("Number of breakdowns")
plt.grid(color='gray',linestyle='--',linewidth=0.3)
#show the result
plt.show()
```
